Загрузка необходимых библиотек

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Загрузка данных

```python
dat=pd.read_excel('C:\YellowBelt/Статусы проектов Данные.xlsx')
dat.head()
```

Подготовка данных для анализа

```python
dat_Jan=dat['Январь'].value_counts()
dat_Nov=dat['Ноябрь'].value_counts()
```

```python
dat_Jan
```

Построение графика по данным января

```python
fig=plt.figure(figsize=(10,8))
ax=fig.add_subplot(111)
ax.pie(dat_Jan,labels=dat_Jan.index,autopct='%1.1f%%',textprops=dict(color='black',size=15))
plt.title('Январь',fontsize=20)
```

Использование разрывов для расставления акцентов

```python
explode=[]
for ind in dat_Jan.index:
    if ind==dat_Jan.idxmax():
        explode.append(0.2)
    else:
        explode.append(0)
explode=tuple(explode)
```

```python
explode
```

```python
fig=plt.figure(figsize=(10,8))
ax=fig.add_subplot(111)
ax.pie(dat_Jan,labels=dat_Jan.index,autopct='%1.1f%%',textprops=dict(color='black',size=15),explode=explode)
plt.title('Январь',fontsize=20)
```

Построение двух графиков

```python
fig=plt.figure(figsize=(10,8))
ax=fig.add_subplot(121)
ax.pie(dat_Jan,labels=dat_Jan.index,autopct='%1.1f%%',counterclock=False,startangle=90)
plt.title('Январь')

ax2=fig.add_subplot(122)
ax2.pie(dat_Nov,labels=dat_Nov.index,autopct='%1.1f%%',counterclock=False,startangle=90)
plt.title('Ноябрь')
```
