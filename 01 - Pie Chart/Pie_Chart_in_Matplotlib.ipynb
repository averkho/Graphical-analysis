Importing all necessary libraries and modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Importing data

```python
dat=pd.read_excel('C:\YellowBelt/Project statuses Data.xlsx')
```

```python
dat.head()
```

Prepare data for analysis

```python
dat_Jan=dat['Status January'].value_counts()
dat_Nov=dat['Status November'].value_counts()
```

```python
dat_Jan
```

Making plot 

```python
fig=plt.figure(figsize=(10,8))
ax=fig.add_subplot(111)
ax.pie(dat_Jan,labels=dat_Jan.index,autopct='%1.1f%%',textprops=dict(color='black',size=15))
plt.title('Project statuses in January',fontsize=20)
```

Using explode to highlight the maximum

```python
explode=[]
for ind in dat_Jan.index:
    if ind==dat_Jan.idxmax():
        explode.append(0.2)
    else:
        explode.append(0)
explode=tuple(explode)
```

```python
explode
```

```python
fig=plt.figure(figsize=(10,8))
ax=fig.add_subplot(111)
ax.pie(dat_Jan,labels=dat_Jan.index,autopct='%1.1f%%',textprops=dict(color='black',size=15),explode=explode)
plt.title('Project statuses in January',fontsize=20)
```

Making two pie charts

```python
fig=plt.figure(figsize=(10,8))
ax=fig.add_subplot(121)

ax.pie(dat_Jan,labels=dat_Jan.index,autopct='%1.1f%%',counterclock=False,startangle=90)
plt.title('Project statuses in January')

ax2=fig.add_subplot(122)
ax2.pie(dat_Nov,labels=dat_Nov.index,autopct='%1.1f%%',counterclock=False,startangle=90)
plt.title('Project statuses in November')
```
